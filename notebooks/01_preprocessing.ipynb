## Chunk Long Dialogues (for MediaSum)

```python
long_text = ' '.join(['This is sentence number ' + str(i) + '.' for i in range(300)])
chunks = chunk_long_dialogue(long_text, max_words=100)
print(f'Total chunks: {len(chunks)}')
for i, c in enumerate(chunks):
    print(f'Chunk {i}: {len(c.split())} words')
```

## Tokenization Check

```python
from transformers import AutoTokenizer
from src.data.preprocessing import tokenize_for_model
from functools import partial

tokenizer = AutoTokenizer.from_pretrained('facebook/bart-large-cnn')

sample_batch = {'dialogue': dataset['train'][:4]['dialogue'],
                'summary':  dataset['train'][:4]['summary']}

fn = partial(tokenize_for_model, tokenizer=tokenizer)
result = fn(sample_batch)
print('Keys:', list(result.keys()))
print('input_ids[0] length:', len(result['input_ids'][0]))
print('labels[0] length:   ', len(result['labels'][0]))
```
