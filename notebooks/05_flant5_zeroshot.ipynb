# Notebook 05 — Flan-T5 Zero-shot
Evaluate google/flan-t5-base with instruction prompting (no fine-tuning).

```python
import sys, json
sys.path.insert(0, '..')

from src.data.loader import load_dialogsum
from src.methods.flant5_zeroshot import FlanT5Summarizer, save_results
```

```python
dataset = load_dialogsum(cache_dir='../data/raw/dialogsum')
test = dataset['test']
dialogues  = test['dialogue']
references = test['summary']
print(f'Test samples: {len(dialogues)}')
```

```python
summarizer = FlanT5Summarizer(model_name='google/flan-t5-base')
print('Model loaded.')
```

```python
# Quick smoke test
sample = dialogues[0]
print('Prediction:', summarizer.summarize(sample))
print('Reference: ', references[0])
```

```python
predictions = summarizer.summarize_batch(dialogues, batch_size=16)
print(f'Generated {len(predictions)} summaries.')
```

```python
scores = save_results(predictions, references, '../results/scores/flant5_results.json')
print('Flan-T5 scores:', scores)
```
